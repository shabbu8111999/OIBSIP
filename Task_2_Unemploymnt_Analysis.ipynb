```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
df.head()
```

**Exploring the Datasets Properly**

```python
# Changing the Column names for better understanding

df.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed',
              'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']
```

**Plotting A Histogram on Estimated Unemployment Rate**

```python
# Visualizing the Estimated Unemployment Rate on the basis of Frequency

plt.figure(figsize=(10, 6))
sns.histplot(df['Estimated Unemployment Rate'], bins = 30, kde = True)
plt.title('Unemployment Rate Distribution')
plt.xlabel('Unemployment Rate')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

**Plotting A Count Plot on the Basis of Estimated Labour Participation Rate**

```python
# Visualizing the Estimated Labour Participation Rate on the basis of their count

plt.figure(figsize=(10, 6))
sns.countplot(x = 'Estimated Labour Participation Rate', data = df[:10], palette = 'rainbow')
plt.title('Estimated Labour Participation Rate')
plt.xlabel('Labour Rate')
plt.ylabel('Frequency')
plt.show()
```

**Displaying A Pie Plot to Show The Regions by Unemploymwnt Rate**

```python
# Creating a Pie Plot for Visualizing the Region by Estimated Unemployment Rate

reg_unemployment = df.groupby('Region')['Estimated Unemployment Rate'].sum()

plt.figure(figsize=(8, 6))
plt.pie(reg_unemployment, labels = reg_unemployment.index, autopct = '%1.1f%%', startangle = 140, shadow = True)
plt.title("Estimated Unemployment Rate by Region")
plt.show()
```

**Visualizing Time Series plot to showcase the Unemployment Rate per month**

```python
# Convert 'Date' to a datetime object
df['Date'] = pd.to_datetime(df['Date'])

fig = px.line(df,
              x = 'Date',
              y = 'Estimated Unemployment Rate',
              title = 'Time Series: Estimated Unemployment Rate',
              labels = {'Estimated Unemployment Rate': 'Unemployment Rate'}
              )
fig.show()
```

**Plotting a Heat Map Correlation of Data**

```python
heat_maps = df.corr()

plt.figure(figsize = (10, 6))
sns.heatmap(data = heat_maps, annot = True, cmap = 'magma', fmt = '.2f')
```

**Plotting a Bar Plot on the basis of Average Unemployment Rate in Each States**

```python
# Visualization of Bar plot by the Average Unemployment Rate in each States

plt_unemployment = df[['Estimated Unemployment Rate', 'States']]

df_unemployed = plt_unemployment.groupby('States').mean().reset_index()

Visual = px.bar(df_unemployed, x = 'Estimated Unemployment Rate',
                y = 'States',
                color = 'States',
                title = 'Average Unemployment Rate in Each State'
                )
Visual.update_layout(autosize = False, width = 1000,  height = 600)

Visual.show()
```

**Displaying the Bar graph for Average Labour Participation Rate in States**

```python
plt_labour = df[['Estimated Labour Participation Rate', 'States']]

df_labour = plt_labour.groupby('States').mean().reset_index()

df_labour = df_labour.sort_values('Estimated Labour Participation Rate')

Visual = px.bar(df_labour, x = 'Estimated Labour Participation Rate',
                y = 'States',
                color = 'States',
                title = 'Average Estimated Labour Participation Rate in Each State'
                )
Visual.update_layout(autosize = False, width = 1000,  height = 650)

Visual.show()
```

**Displaying the Bar graph for Average Labour Participation Rate in Regions**

```python
labour_region = df[['Estimated Labour Participation Rate', 'Region']]

df_lab_reg = labour_region.groupby('Region').mean().reset_index()

Visual = px.bar(df_lab_reg, x = 'Region',
                y = 'Estimated Labour Participation Rate',
                color = 'Region',
                title = 'Average Estimated Labour Participation Rate in Each Region'
                )
Visual.update_layout(width = 800,  height = 650)
Visual.show()
```

**Plotting a Bar Graph for Estimated Employed by Region**

```python
Est_Employed = df[['Estimated Employed', 'Region']]

df_est_empl = Est_Employed.groupby('Region').mean().reset_index()

Visual = px.bar(df_est_empl, x = 'Region',
                y = 'Estimated Employed',
                color = 'Region',
                title = 'Average Estimated Employed in Each Region'
                )
Visual.update_layout(width = 800,  height = 650)
Visual.show()
```

**A Pie Plot on Unemployment Rate in Regions and States**

```python
# Created a Reg_State for sunburst chart
df['Reg_State'] = df['Region'] + ' = ' + df['States']

# Created a sunburst chart
fig = px.sunburst(df, path = ['Reg_State'], values = 'Estimated Unemployment Rate',
                  title = 'Region and States', color = 'Reg_State')

# Updated the layout for better readability
fig.update_layout(margin = dict(l = 0, r = 0, b = 0, t = 40))

fig.show()
```
